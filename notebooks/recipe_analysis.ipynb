# List of Ingredients(from Flavor DB) that have recipes on Recipe Puppy API

```python
#Opening the pickled file
pickle_in = open("../data/ingredients/ingredient_full_list.pickle","rb")

#Getting the dictionary from the pickle
pickled_list = pickle.load(pickle_in)
```

```python
#Making each ingredient name lowercase
lowered_pickle = [x.lower() for x in pickled_list]
```

```python
#Making the list into a set for comparison
ingredient_set = set(lowered_pickle)
ingredient_set
```

```python
print("number of ingredients that have a recipe in Recipe Puppy API: ", len(ingredient_set))
```

# PandaDF of Recipe and associated Ingredients from Recipe Puppy API

```python
#accessing mongoDB
client = MongoClient()
database = client['food_map']   # Database name (to connect to)
collections = database['recipe']

#Getting the dataset from MongoDB into Pandas
recipe_puppy_pandas = pd.DataFrame(list(collections.find()))
```

```python
recipe_puppy_pandas.shape
```

```python
recipe_puppy_pandas.head()
```

```python
# Dropping recipes that have the same name that will be duplicates
recipe_puppy_pandas = recipe_puppy_pandas.drop_duplicates(subset='recipe_name', keep="last")
```

```python
recipe_puppy_pandas.shape
```

```python
with open('../data/pandas/recipe_puppy_pandas.pickle', 'wb') as file:
    file.write(pickle.dumps(recipe_puppy_pandas))
    file.close()
```

# All the ingredients from the recipes in Recipe Puppy API

```python
# iterate through each recipe
for index, row in recipe_puppy_pandas.iterrows():   
    ingredient_list = row["recipe_ingredients"]
    print(ingredient_list)
    break
```

```python
recipe_set = set(recipe_puppy_pandas["searched_ingredient"].unique())
recipe_set
```

```python
non_used_ingredients = ingredient_set.difference(recipe_set)
non_used_ingredients
```

```python
with open('non_used_ingredients.pickle', 'wb') as file:
    file.write(pickle.dumps(non_used_ingredients))
    file.close()
```

```python
len(non_used_ingredients)
```

```python
used_ingredients = ingredient_set.intersection(recipe_set)
used_ingredients
```

```python
with open('ingredients_with_recipes.pickle', 'wb') as file:
    file.write(pickle.dumps(used_ingredients))
    file.close()
```
